### Punto 3 - Regresión lineal
En la celda 3 ajuste un modelo de regresión lineal a un conjunto de train de datos, utilizando "total" como variable de respuesta y "season" y "hour" como las únicas variables predictoras, teniendo en cuenta que la variable "season" es categórica. Luego, imprima los coeficientes e interprételos. ¿Cuáles son las limitaciones de la regresión lineal en este caso?

```python
# Celda 3
# Separar variables predictoras y target
X = bikes[['hour', 'season']]
y = bikes['total']

corregir = {1:'invierno',2:'primavera',3:'verano',4:'otoño'}
X['season'] = X['season'].map(corregir)

# Aplicar One-Hot Encoding 
X_encoded = X.copy()
season_encoded = pd.get_dummies(X['season'], drop_first = True)
X_encoded = pd.concat([X['hour'], season_encoded], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=1)

# Modelo de regresión lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X_train, y_train)

# Obtener los coeficientes del modelo ajustado
coeficientes = modelo_regresion.coef_
intercepto = modelo_regresion.intercept_

# Imprimir coeficientes
for i in range(len(X_train.columns)):
  print(f"Coeficiente de {X_train.columns[i]}: {coeficientes[i]}")

# Imprimir intercepto
print(f"Intercepto de la regresión: {intercepto}")

# Obtener las predicciones con datos de test
pred_regresion = modelo_regresion.predict(X_test)

# Evaluar el modelo con datos de test
mse_regresion = mean_squared_error(y_test, pred_regresion)

print(f"Evaluación del modelo en test MSE: {mse_regresion}")
```

#### Conclusión regresión lineal

**Interpretación de coefcientes**


Al imprimir los coeficientes, se observa que el coeficiente de la variable hour es de 10.628,  lo cual indica que, según el modelo, por cada incremento de una unidad en la hora, la cantidad de rentas de bicicleta aumenta en aproximadamente 11 unidades. Sin embargo, esta relación lineal podría no ser la más adecuada, ya que, como se observó en el análisis preliminar, la relación entre hour y las rentas de bicicleta no es lineal. Se detectó un patrón en el que las rentas aumentan hasta cierto punto, luego disminuyen y vuelven a aumentar, lo cual no puede ser capturado adecuadamente por un modelo de regresión lineal.

En cuanto a las variables de estaciones (otoño, primavera y verano), el coeficiente de otoño es de 81.14, mientras que el de primavera es de 93.83 y el de verano es de 118.74, lo cual coincide con las conclusiones a las que se llegaron en el análisis preliminar en donde se observaba el siguiente orden en la media de rentas de bicicleta por hora en las diferentes estaciones:
1. verano (media más alta)
2. primavera
3. otoño
4. invierno

**Limitantes de Regresión Lineal**

Una de las principales limitaciones de la regresión lineal en este caso es su incapacidad para modelar relaciones no lineales entre las variables predictoras y la variable objetivo. Como se mencionó, la relación entre la variable hour y la cantidad de rentas de bicicletas no es lineal, lo que significa que un modelo de regresión lineal no puede capturar adecuadamente los picos y valles observados en los datos. Además, la regresión lineal supone que las relaciones entre las variables son constantes a lo largo de todo el rango de los datos, lo cual no se ajusta a la realidad en muchos casos.

### Punto 4 - Árbol de decisión manual
En la celda 4 cree un árbol de decisiones para pronosticar la variable "total" iterando **manualmente** sobre las variables "hour" y  "season". El árbol debe tener al menos 6 nodos finales.

```python
# Celda 4

def mse(y):
    return ((y - y.mean()) ** 2).mean()

def mse_impurity(X_col, y, split):
    filter_l = X_col < split if X_col.dtype != 'O' else X_col == split
    y_l = y.loc[filter_l]
    y_r = y.loc[~filter_l]

    n_l = y_l.shape[0]
    n_r = y_r.shape[0]

    if n_l == 0 or n_r == 0:
        return 0

    mse_y = mse(y)
    mse_l = mse(y_l)
    mse_r = mse(y_r)

    # Ganancia basada en la reducción de MSE
    mse_impurity_ = mse_y - (n_l / (n_l + n_r) * mse_l + n_r / (n_l + n_r) * mse_r)

    return mse_impurity_

# puntos de corte seleccionados arbitrariamente
def best_split(X, y):
    '''
        Funcion que al pasarle las X y y itera sobre las columnas de X (features) y
        retorna una lista de best_split en donde el
        primer elemento es el feature que uso para el split
        el segundo elemento es el split o el valor en donde dividio y
        el tercer elemento es el gain de haberlo hecho
    '''
    features = X.columns
    best_split = [None, None, 0]  # feature, split, gain

    for feature in features:
        if X[feature].dtype == 'O' or X[feature].nunique() < 5:
            # usa los valores unicos como posibles puntos de division
            splits = X[feature].unique()
        else:
            # si es numerica agarra puntos de corte en cada hora
            min_val = X[feature].min()
            max_val = X[feature].max()
            splits = np.arange(min_val, max_val+1)

        for split in np.unique(splits):
            gain = mse_impurity(X[feature], y, split)
            if gain > best_split[2]:
                best_split = [feature, split, gain]

    return best_split

def split_data(X, y, feature, value):
    '''
      Función que divide los datos en base a una característica y un valor de corte
      se utiliza dentro de la función de crecer arbol para dividir los datos de acuerdo al best split
      segun la funcion de best_split
    '''
    if X[feature].dtype == 'O' or X[feature].nunique() < 5:  # Si es categórica
        left_mask = X[feature] == value
    else:  # Si es numérica
        left_mask = X[feature] < value

    return X[left_mask], y[left_mask], X[~left_mask], y[~left_mask]

def grow_tree(X, y, depth=0, max_depth=3, min_leaves=6, leaf_count=[0]):
    '''
      FUNCION RECURSIVA que divide los datos hasta que:
      se alcanza la maxima profundidad definida
      o hasta que ya no se puede dividir mas los datos (y es pequeño o solamente tiene una categoria)
      o hasta que ya no hay ganancia del corte que se hizo
    '''

    # Si alcanzamos la profundidad máxima, convertimos el nodo en hoja y se aumenta el conteo de nodos terminales
    if depth >= max_depth:
        leaf_count[0] += 1
        return {"prediction": y.mean()}

    # Si ya hay suficientes hojas y no se puede dividir más igual se convierte en hoja y se aumenta el conteo
    if len(y) < 2 or len(set(y)) == 1:
        if leaf_count[0] >= min_leaves:
            leaf_count[0] += 1
            return {"prediction": y.mean()}

    # si no ocurrio ninguno de lo anterior, se intenta encontrar el best split con los datos
    best_feature, best_split_val, gain = best_split(X, y)

    # Si no hay ganancia, se para
    if gain == 0:
        leaf_count[0] += 1
        return {"prediction": y.mean()}

    # pero si si hay ganancia, se dividien los datos de acuerdo al split
    X_left, y_left, X_right, y_right = split_data(X, y, best_feature, best_split_val)

    # y se llama la funcion de cada lado con los datos divididos
    # SE MANDA LLAMAR LA FUNCION DENTRO DE LA FUNCION
    left_tree = grow_tree(X_left, y_left, depth + 1, max_depth, min_leaves, leaf_count)
    right_tree = grow_tree(X_right, y_right, depth + 1, max_depth, min_leaves, leaf_count)

    return {
        "feature": best_feature,
        "split": best_split_val,
        "left": left_tree,
        "right": right_tree,
    }
tree = grow_tree(bikes[["hour", "season"]], bikes["total"], min_leaves=6)

def print_tree(tree, depth=0):
    """Imprime el árbol de decisión con indentación."""
    if "prediction" in tree:
        print("  " * depth + f"🌿 Predicción: {tree['prediction']:.2f}")
    else:
        print("  " * depth + f"Si {tree['feature']} < {tree['split']}:")
        print_tree(tree["left"], depth + 1)
        print("  " * depth + f"Si {tree['feature']} >= {tree['split']}:")
        print_tree(tree["right"], depth + 1)

print_tree(tree)
```

### Punto 5 - Árbol de decisión con librería
En la celda 5 entrene un árbol de decisiones con la **librería sklearn**, usando las variables predictoras "season" y "hour" y calibre los parámetros que considere conveniente para obtener un mejor desempeño. Recuerde dividir los datos en conjuntos de entrenamiento y validación para esto. Comente el desempeño del modelo con alguna métrica de desempeño de modelos de regresión y compare desempeño con el modelo del punto 3.

```python
# Celda 5
from sklearn.model_selection import RandomizedSearchCV

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=1)


# Definimos el rango de los hiperparámetros a buscar
param_dist = {
    'max_depth': range(1, 21),  # Maximo nivel de profundidad
    'min_samples_split': range(2, 21),  # Minimo número de muestras para dividir un nodo
    'min_samples_leaf': range(1, 21)  # Minimo número de muestras en un nodo hoja
}

# Definir el modelo
arbol = DecisionTreeRegressor(random_state=1)

# Configurar RandomizedSearchCV
random_search = RandomizedSearchCV(
    arbol,
    param_distributions=param_dist,
    n_iter=10,  # Número de combinaciones aleatorias a probar
    cv=10,  # Validación cruzada con 10 particiones
    scoring='neg_mean_squared_error',  # Queremos minimizar el MSE
    random_state=1,
)

# Ajustar el modelo con los datos
random_search.fit(X_train, y_train)

# Imprimir los mejores parámetros encontrados y el mejor puntaje
print("Mejores parámetros:", random_search.best_params_)

# Evaluar el modelo con el conjunto de prueba
best_model = random_search.best_estimator_
print(best_model)
y_pred = best_model.predict(X_test)

# Calcular el MSE en el conjunto de prueba
mse_test = mean_squared_error(y_test, y_pred)
print(f'MSE en el conjunto de prueba: {mse_test}')
```

#### Conclusión y comparación

El MSE del conjunto de prueba para el árbol de decisión es de aproximadamente 13,225, mientras que para la regresión lineal es de alrededor de 26,473. Esto indica que el árbol de decisión ofrece un mejor desempeño en la predicción en este caso, ya que su error cuadrático medio es significativamente menor. Esto podría deberse a que los árboles de decisión pueden capturar de mejor manera relaciones no lineales en los datos, mientras que la regresión lineal asume una relación estrictamente lineal entre las variables.

### Datos popularidad de noticias

```python
# Lectura de la información de archivo .csv
df = pd.read_csv('https://raw.githubusercontent.com/davidzarruk/MIAD_ML_NLP_2023/main/datasets/mashable.csv', index_col=0)
df.head()
```

```python
# División de la muestra en set de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

### Punto 6 - Árbol de decisión y regresión logística
En la celda 6 construya un árbol de decisión y una regresión logística. Para el árbol calibre al menos un parámetro y evalúe el desempeño de cada modelo usando las métricas de Accuracy y F1-Score.

```python
# Celda 6
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# arboles calibrando max_depth y min_samples_leaf
# Definimos el rango de los hiperparámetros a buscar
param_dist = {
    'max_depth': range(1, 21),  # Maximo nivel de profundidad
    'min_samples_leaf': range(1, 21)  # Minimo número de muestras en un nodo hoja
}

# Definir el modelo
arbol_popularidad = DecisionTreeClassifier(random_state=1)

# Configurar RandomizedSearchCV
random_search = RandomizedSearchCV(
    arbol_popularidad,
    param_distributions=param_dist,
    n_iter=10,  # Número de combinaciones aleatorias a probar
    cv=10,  # Validación cruzada con 10 particiones
    scoring='accuracy',  # Queremos maximizar accuracy
    random_state=1,
    n_jobs=-1  # Para usar todos los cores disponibles
)

# Ajustar el modelo con los datos
random_search.fit(X_train, y_train)

# Imprimir los mejores parámetros encontrados y el mejor puntaje
print("Mejores parámetros:", random_search.best_params_)

# Evaluar el modelo con el conjunto de prueba
best_arbol_popularidad = random_search.best_estimator_
predicciones_best_arbol = best_arbol_popularidad.predict(X_test)

accuracy_best_arbol = metrics.accuracy_score(predicciones_best_arbol, y_test)
f1_best_arbol = metrics.f1_score(predicciones_best_arbol, y_test)

print(f"F1-score de arboles con max_depth determinado: {f1_best_arbol}")
print(f"Accuracy de arboles con max_depth determinado: {accuracy_best_arbol}")
```

```python
# Regresion logistica
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
predicciones_lr = lr.predict(X_test)

f1_lr = metrics.f1_score(predicciones_lr, y_test)
accuracy_lr = metrics.accuracy_score(predicciones_lr, y_test)

print(f"F1-score de arboles con max_depth determinado: {f1_lr}")
print(f"Accuracy de arboles con max_depth determinado: {accuracy_lr}")
```

#### Conclusión Árbol vs Regresión Logística

Al igual que con el dataset anterior, los árboles demuestran un mejor desempeño, de manera similar, se puede concluir que los árboles logran capaturar de mejor manera relaciones no lineales entre las variables, lo cual puede resultar en un  mejor modelo.

### Punto 7 - Votación Mayoritaria
En la celda 7 elabore un esamble con la metodología de **Votación mayoritaria** compuesto por 300 muestras bagged donde:

-las primeras 100 muestras vienen de árboles de decisión donde max_depth tome un valor de su elección\
-las segundas 100 muestras vienen de árboles de decisión donde min_samples_leaf tome un valor de su elección\
-las últimas 100 muestras vienen de regresiones logísticas

Evalúe cada uno de los tres modelos de manera independiente utilizando las métricas de Accuracy y F1-Score, luego evalúe el ensamble de modelos y compare los resultados. 

Nota: 

Para este ensamble de 300 modelos, deben hacer votación mayoritaria. Esto lo pueden hacer de distintas maneras. La más "fácil" es haciendo la votación "manualmente", como se hace a partir del minuto 5:45 del video de Ejemplo práctico de emsablajes en Coursera. Digo que es la más fácil porque si hacen la votación mayoritaria sobre las 300 predicciones van a obtener lo que se espera.

Otra opción es: para cada uno de los 3 tipos de modelos, entrenar un ensamble de 100 modelos cada uno. Predecir para cada uno de esos tres ensambles y luego predecir como un ensamble de los 3 ensambles. La cuestión es que la votación mayoritaria al usar los 3 ensambles no necesariamente va a generar el mismo resultado que si hacen la votación mayoritaria directamente sobre los 300 modelos. Entonces, para los que quieran hacer esto, deben hacer ese último cálculo con cuidado.

Para los que quieran hacerlo como ensamble de ensambles, digo que se debe hacer el ensamble final con cuidado por lo siguiente. Supongamos que:

* para los 100 árboles del primer tipo, la votación mayoritaria es: 55% de los modelos predicen que la clase de una observación es "1"
* para los 100 árboles del segundo tipo, la votación mayoritaria es: 55% de los modelos predicen que la clase de una observación es "1"
* para las 100 regresiones logísticas, la votación mayoritaria es: 10% de los modelos predicen que la clase de una observación es "1"

Si se hace la votación mayoritaria de los 300 modelos, la predicción de esa observación debería ser: (100*55%+100*55%+100*10%)/300 = 40% de los modelos votan porque la predicción debería ser "1". Es decir, la predicción del ensamble es "0" (dado que menos del 50% de modelos predijo un 1).

Sin embargo, si miramos cada ensamble por separado, el primer ensamble predice "1", el segundo ensamble predice "1" y el último ensamble predice "0". Si hago votación mayoritaria sobre esto, la predicción va a ser "1", lo cual es distinto a si se hace la votación mayoritaria sobre los 300 modelos.

```python
# Celda 7

# Creación de 300 muestras de bootstrap
np.random.seed(123)
# tomando la cantidad de datos que hay en X_train
n_observaciones = X_train.shape[0]
n_estimators = 300

# list comprehension de listas con np.random.choice
samples = [np.random.choice(a=n_observaciones, size=n_observaciones, replace=True) for _ in range(1, n_estimators +1 )]
print(f"Bootstrap samples: {len(samples)}")

sample_arboles_1 = samples[0:100]
sample_arboles_2 = samples[100:200]
sample_regresionlog = samples[200:300]

print(f"Samples para arboles con max_depth: {len(sample_arboles_1)} cada un con {len(sample_arboles_1[0])} observaciones")
print(f"Samples para arboles con min_samples_leaf: {len(sample_arboles_2)} cada un con {len(sample_arboles_2[0])} observaciones")
print(f"Samples para modelos de regresion logistica: {len(sample_regresionlog)} cada una con {len(sample_regresionlog[0])} observaciones")
```

#### Arboles 1

```python
# creacion de primeros 100 estimators (arboles con max depth)
n_estimators = len(sample_arboles_1)
np.random.seed(123)
seeds = np.random.randint(1, 10000, size=n_estimators)

trees_1 = {}

for i in range(n_estimators):
    # escogi max_depth = 5
    trees_1[i] = DecisionTreeClassifier(max_depth=5, random_state=seeds[i])
    trees_1[i].fit(X_train.iloc[sample_arboles_1[i]], y_train.iloc[sample_arboles_1[i]])

accuracies_individuales1 = []

# Predicción para los datos del set de test con cada modelo
y_pred_df_1 = pd.DataFrame(index=X_test.index, columns=list(range(n_estimators)))
for i in range(n_estimators):
    y_pred_df_1.iloc[:, i] = trees_1[i].predict(X_test)

    accuracy_arboles1 = metrics.accuracy_score(trees_1[i].predict(X_test), y_test)
    accuracies_individuales1.append(accuracy_arboles1)

y_pred_df_1.head()
```

#### Arboles 2

### Punto 8 - Votación Ponderada
En la celda 8 elabore un ensamble con la metodología de **Votación ponderada** compuesto por 300 muestras bagged para los mismos tres escenarios del punto 7. Evalúe los modelos utilizando las métricas de Accuracy y F1-Score

```python
# Celda 8
samples_oob = []
n_samples = X_train.shape[0]

# Obtención de las observaciones fuera de la bolsa "out-of-bag" para cada muestra
for sample in samples:
    samples_oob.append(sorted(set(range(n_samples)) - set(sample)))

n_estimators = 300

# Estimación de los errores OOB para cada clasificador
errors = np.zeros(n_estimators)

for i in range(n_estimators):
    if i >= 0 and i <100:
      y_pred_ = trees_1[i].predict(X_train.iloc[samples_oob[i]])
      errors[i] = 1 - metrics.accuracy_score(y_train.iloc[samples_oob[i]], y_pred_)
    if i >= 100 and i<200:
      y_pred_ = trees_2[i-100].predict(X_train.iloc[samples_oob[i]])
      errors[i] = 1 - metrics.accuracy_score(y_train.iloc[samples_oob[i]], y_pred_)
    if i >= 200 and i<300:
      y_pred_ = logisticregressors[i-200].predict(X_train.iloc[samples_oob[i]])
      errors[i] = 1 - metrics.accuracy_score(y_train.iloc[samples_oob[i]], y_pred_)
```

```python
# Visualización de OOB para cada árbol

plt.scatter(range(n_estimators), errors)
plt.xlim([0, n_estimators])
plt.title('OOB Errors of each tree')
```

```python
# Obtención de los pesos alpha de cada modelo de acuerdo al error OOB
# alpha = accuracies / accuracies.sum()

alpha = (1 - errors) / (1 - errors).sum()

lista_alphas = pd.Series(alpha)
lista_alphas.plot()
plt.title('Gráfica de Alphas')
```

```python
# Ponderación de las predicciones con los pesos alpha
df_ponderada = estimators_300.iloc[:,0:300]
df_ponderada.head()
```

```python
df_multiplicada = df_ponderada.copy()
df_multiplicada = (df_multiplicada * alpha)
df_multiplicada['suma_ponderada'] = df_multiplicada.sum(axis=1)
df_multiplicada['pred voto ponderado'] = (df_multiplicada['suma_ponderada'] >= 0.5).astype(np.int64)
df_multiplicada.head()
```

```python
f1_ponderado = metrics.f1_score(df_multiplicada['pred voto ponderado'], y_test)
accuracy_ponderado = metrics.accuracy_score(df_multiplicada['pred voto ponderado'], y_test)

print(f"F1-score de ensamble 300 modelos - voto ponderado: {f1_ponderado}")
print(f"Accuracy de ensamble 300 modelos - voto ponderado: {accuracy_ponderado}")
```

#### Conclusión Votación Ponderada

En este caso, no se muestra una mejora relevante entre ambos tipos de votaciones en este ensamblaje, pues las diferencias entre los F1-scores y accuracies son mínimas.

El enfoque de realizar múltiples muestras a través de bootstrapping y entrenar diferentes tipos de modelos (árboles con max_depth, árboles con min_samples_leaf y regresión logística) es una estrategia válida para explorar diferentes configuraciones y obtener modelos diversos. Los ensambles de estos modelos, especialmente cuando se combinan en grupos diversos, pueden mejorar el rendimiento, ya que aprovechan la complementariedad de los errores de los modelos individuales.

En este caso, el ensamble de modelos con `min_samples_leaf` determinado tuvo un rendimiento destacado, posiblemente debido a la mayor robustez y la diversidad en los errores entre los modelos. Sin embargo, los ensambles de los 300 modelos no mostró una mejora significativa, lo que puede explicarse por la redundancia en las configuraciones, la pérdida de diversidad y el riesgo de sobreajuste que puede surgir al incluir demasiados modelos similares. La clave para un buen ensamble no está únicamente en la cantidad de modelos, sino en asegurar que estos sean lo suficientemente diversos como para aprovechar la compensación de errores y mejorar la capacidad de generalización.

**Ventajas y desventajas**

Las ventajas de utilizar ensambles radican en la capacidad de complementar los errores de modelos individuales y combinar diferentes tipos de modelos, como en este caso, al juntar árboles y regresiones para aprovechar la diversidad que cada uno ofrece. Sin embargo, una desventaja es que para que un ensamble mejore el rendimiento, debe existir diversidad entre los modelos; si estos son demasiado similares, no se complementan adecuadamente y la efectividad del ensamble disminuye.
